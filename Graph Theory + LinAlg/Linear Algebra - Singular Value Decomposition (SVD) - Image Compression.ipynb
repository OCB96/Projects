```python
theta = np.pi/2
rotationalmatrix = np.array([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]])
```

```python
plt.figure(figsize = (5,5))
x = 12
y = 2
pos = [x,y]

plt.plot([0, pos[0]],[0, pos[1]], label = 'original line')

dotproduct = np.dot(rotationalmatrix, pos)
plt.plot([0, dotproduct[0]],[0, dotproduct[1]], label = 'rotated line')
plt.legend()
shayp = np.shape(dotproduct)
print (shayp)
shayptwo = np.shape(rotationalmatrix)
print (shayptwo)
```

```python
housematrix = np.array([[-6,-6,-7,0,7,6,6,-3,-3,0,0,-6],[-7,2,1,8,1,2,-7,-7,-2,-2,-7,-7]])
```

```python
# i = 0
# j = 0

# for i in housematrix[i][j]:
#     for j in housematrix [i][j]:
#         plt.plot([0, housematrix[i]], [0, housematrix[j]])
dotproductofhouse = np.dot(rotationalmatrix, housematrix)
inverseofrotationalmatrix = np.linalg.inv(rotationalmatrix)
newdp = np.dot(inverseofrotationalmatrix, dotproductofhouse)
plt.subplot(2,2,1)
plt.plot(housematrix[0,:], housematrix[1,:])
plt.subplot(2,2,2)
plt.plot(dotproductofhouse[0,:], dotproductofhouse[1,:])
plt.subplot(2,2,3)
plt.plot(newdp[0,:], newdp[1,:])
```

```python
hedy = mpimg.imread('hedy.jpg')
shep = np.shape(hedy)
print (shep)

plt.figure(figsize=(8,8))
plt.subplot(2,2,1)

actualstoragespace = np.size(hedy)
plt.title('%d' % actualstoragespace)

plt.imshow(hedy, cmap='gray')
plt.axis('off')
U, S, V = np.linalg.svd(hedy)
plt.subplot(2,2,2)
plt.plot(S)
```
