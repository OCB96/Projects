```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G1 = nx.Graph()
elist = [(1, 2), (2, 3), (1, 4), (2, 5), (2, 6), (3, 7), (4, 5), (5, 6), (6, 7), (5, 8), (6, 9), (8, 9)]
G1.add_edges_from(elist)
nx.draw(G1)
plt.show()

A = nx.to_scipy_sparse_matrix(G1, nodelist = [1, 2, 3, 4, 5, 6, 7, 8, 9])
A[1].nnz
A[2].nnz
A[3].nnz
A[4].nnz
A[5].nnz
A[6].nnz
A[7].nnz
A[8].nnz
A[9].nnz
```
