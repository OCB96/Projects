```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
## a graph of order 4 size 5
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(2, 4)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 1)

plt.subplot(121)
nx.draw(G, pos=nx.circular_layout(G), nodecolor='r', edge_color='b')
plt.show()
```

```python
## an order 3 digraph of my choice

G = nx.DiGraph()
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 2)
G.add_edge(2, 1)

plt.subplot(122)
nx.draw(G, pos=nx.circular_layout(G), nodecolor='r', edge_color='b')
plt.show()
```
