```python
import collections
import scipy.stats
```

```python
n, p = 100, 0.02
g3 = nx.gnp_random_graph(n,p)
nx.draw(g3, with_labels = True, node_color = 'y')
plt.show()
deg = nx.degree(g3)
print(deg)
```

```python
dseq = sorted(nx.degree(g3).values(), reverse = True)
dcount = collections.Counter(dseq)
deg, cnt = zip(*dcount.items())

#plot histogram
fig, ax = plt.subplots()
plt.bar(deg, cnt, width = 0.80, color = 'b')
plt.title("degree histogram")
plt.ylabel("count")
plt.xlabel("degree")
ax.set_xticks([d for d in deg])
ax.set_xticklabels(deg)
plt.show()

#plot distribution
fig, ax = plt.subplots()
plt.bar(deg, cnt/(np.sum(cnt)), width = 0.80, color = 'y')
plt.title("degree distribution")
plt.ylabel("count")
plt.xlabel("degree")
ax.set_xticks([d for d in deg])
ax.set_xticklabels(deg)
plt.show()
```

```python
n, p = 100, 0.02
g4 = nx.gnp_random_graph(n,p)
dseq = sorted(nx.degree(g4).values(), reverse = True)
dcount = collections.Counter(dseq)
deg, cnt = zip(*dcount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt/(np.sum(cnt)), width = 0.80, color = 'g')
plt.title("degree distribution")
plt.ylabel("probability")
plt.xlabel("degree")
ax.set_xticks([d for d in deg])
ax.set_xticklabels(deg)
x = scipy.linspace(0, 10, 11)
pmf = scipy.stats.binom.pmf(x, n, p)
plt.plot(x, pmf)


# draw a graph in insets
plt.axes([0.4, 0.4, 0.5, 0.5])
gcc = sorted(nx.connected_component_subgraphs(g4), key=len, reverse = True)
pos = nx.circular_layout(g4)
plt.axis('off')
nx.draw_networkx_nodes(g4, pos, node_size = 20, node_color = 'y')
nx.draw_networkx_edges(g4, pos, alpha = 0.4, edge_color = 'y')
plt.show()
```
